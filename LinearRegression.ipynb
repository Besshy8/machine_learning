```python
import pandas as pd
from sklearn import datasets
```

```python
boston = datasets.load_boston()
```

```python
df = pd.DataFrame(data=boston.data,columns=boston.feature_names)
df.head()
```

```python
df['value'] = boston.target
```

```python
df.head()
```

```python
df.loc[:,'CRIM':'value'].corr()
```

```python
import seaborn as sns
sns.set_style('whitegrid')
```

```python
sns.pairplot(df.loc[:,'CRIM':'value'])
```

```python
import numpy as np
cm = np.corrcoef(df.values.T)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,13))
sns.heatmap(cm,
           cmap='Blues',
           annot=True,
           annot_kws={'size':20},
           fmt='1.1f',
           xticklabels= df.loc[:,'CRIM':'value'].columns,
           yticklabels=df.loc[:,'CRIM':'value'].columns)
```

```python
X = df[["RM"]].values
```

```python
y = df[["value"]].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
```

```python
slr.fit(X_train,y_train)
```

```python
pred = slr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test,pred))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X_pca = pca.fit_transform(df.loc[:,'CRIM':'LSTAT']) ##X_pcanumpy
df_pca = pd.DataFrame(X_pca,columns=['1st','2nd','3rd','4th','5th','6th','7th','8th','9th','10th','11th','12th','13th'])
df_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.hstack([0,pca.explained_variance_ratio_.cumsum()]))
plt.show()
```

```python
pca.components_
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(pca.components_,
           cmap='Blues',
           annot=True,
           annot_kws={'size':10},
           fmt='1.1f',
           xticklabels= df.loc[:,'CRIM':'LSTAT'].columns,
           yticklabels=['1st','2nd','3rd','4th','5th','6th','7th','8th','9th','10th','11th','12th','13th'])
```

```python
df_pca['value']=df['value']

fig,ax = plt.subplots(figsize=(10,6))
df_pca.plot(kind = 'scatter',x='1st',y='2nd',s=100,c='value',cmap='winter',alpha=0.5,ax = ax)
```
